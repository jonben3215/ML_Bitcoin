```python
#Imports
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read Data
data = pd.read_csv('/home/jovyan/ML_Bitcoin/bitcoin_2017_to_2023.csv')
```

```python
#Data Cleaning

#Handling Missing Values
data.isnull().sum()
data.dropna()

#Removing Duplicates
data.duplicated().sum()
data.drop_duplicates()
```

```python
#Outliers
O_Q1 = data['open'].quantile(0.25)
O_Q3 = data['open'].quantile(0.75)
O_IQR = O_Q3 - O_Q1

OLow = O_Q1 - 1.5 * O_IQR
OHigh = O_Q3 + 1.5 * O_IQR

O_Outliers = data[(data['open'] < OLow ) | (data['open'] > OHigh)]

Remove_O = data[(data['open'] >= OLow) & (data['open'] <= OHigh)]

#Outliers - High
H_Q1 = data['high'].quantile(0.25)
H_Q3 = data['high'].quantile(0.75)
H_IQR = H_Q3 - H_Q1

HLow = H_Q1 - 1.5 * H_IQR
HHigh = H_Q3 + 1.5 * H_IQR

# H_Outliers = data[(data['high'] < HLow ) | (data['high'] > HHigh)]

Remove_H = data[(data['high'] >= HLow) & (data['high'] <= HHigh)]
```

```python
#Outliers - Low
L_Q1 = data['low'].quantile(0.25)
L_Q3 = data['low'].quantile(0.75)
L_IQR = L_Q3 - L_Q1

LLow = L_Q1 - 1.5 * L_IQR
LHigh = L_Q3 + 1.5 * L_IQR

L_Outliers = data[(data['low'] < LLow ) | (data['low'] > LHigh)]

Remove_L = data[(data['low'] >= LLow) & (data['low'] <= LHigh)]
#Outliers - Close
C_Q1 = data['close'].quantile(0.25)
C_Q3 = data['close'].quantile(0.75)
C_IQR = C_Q3 - C_Q1

CLow = C_Q1 - 1.5 * C_IQR
CHigh = C_Q3 + 1.5 * C_IQR

# C_Outliers = data[(data['close'] < CLow ) | (data['close'] > CHigh)]

Remove_C = data[(data['close'] >= CLow) & (data['close'] <= CHigh)]


```

```python
# Plot it 
#high	low	close	volume	quote_asset_volume	
#number_of_trades	taker_buy_base_asset_volume
#taker_buy_quote_asset_volume
#Before Outlier Reduction
plt.boxplot(data['open'])
plt.title('Before Handling Outliers: OPEN')
plt.show()
#After Outlier Reduction
plt.boxplot(Remove_O['open'])
plt.title('After Handling Outliers: OPEN')
plt.show()

plt.boxplot(data['low'])
plt.title('Before Handling Outliers: LOW')
plt.show()

#After Outlier Reduction
plt.boxplot(Remove_L['low'])
plt.title('After Handling Outliers: LOW')
plt.show()

plt.boxplot(data['high'])
plt.title('Before Handling Outliers: HIGH')
plt.show()
#After Outlier Reduction
plt.boxplot(Remove_H['high'])
plt.title('After Handling Outliers: HIGH')
plt.show()

plt.boxplot(data['close'])
plt.title('Before Handling Outliers: CLOSE')
plt.show()

#After Outlier Reduction
plt.boxplot(Remove_C['close'])
plt.title('After Handling Outliers: CLOSE')
plt.show()

# plt.boxplot(data['volume'])
# plt.title('Before Handling Outliers: VOLUME')
# plt.show()
# #After Outlier Reduction
# plt.boxplot(Remove_V['volume'])
# plt.title('After Handling Outliers: VOLUME')
# plt.show()

# # plt.boxplot(data['number_of_trades'])
# # plt.title('Before Handling Outliers: NUMBER OF TRADES')
# # plt.show()

# # #After Outlier Reduction
# # plt.boxplot(Remove_NOT['number_of_trades'])
# # plt.title('After Handling Outliers: NUMBER OF TRADES')
# # plt.show()
# # quote_asset_volume	
# #number_of_trades	taker_buy_base_asset_volume
# #taker_buy_quote_asset_volume
# plt.boxplot(data['quote_asset_volume'])
# plt.title('Before Handling Outliers: QUOTE ASSET VOLUME')
# plt.show()
# #After Outlier Reduction
# plt.boxplot(Remove_QAV['quote_asset_volume'])
# plt.title('After Handling Outliers: QUOTE ASSET VOLUME')
# plt.show()

# plt.boxplot(data['taker_buy_base_asset_volume'])
# plt.title('Before Handling Outliers: TAKER BUY BASE ASSET VOLUME')
# plt.show()

# #After Outlier Reduction
# plt.boxplot(Remove_TBBAV['taker_buy_base_asset_volume'])
# plt.title('After Handling Outliers: TAKER BUY BASE ASSET VOLUME')
# plt.show()

# plt.boxplot(data['taker_buy_quote_asset_volume'])
# plt.title('Before Handling Outliers: TAKER BUY QUOTE ASSET VOLUME')
# plt.show()

# #After Outlier Reduction
# plt.boxplot(Remove_TBQAV['taker_buy_quote_asset_volume'])
# plt.title('After Handling Outliers: TAKER BUY QUOTE ASSET VOLUME')
# plt.show()
```
