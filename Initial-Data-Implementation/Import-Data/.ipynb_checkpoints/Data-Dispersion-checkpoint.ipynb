```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.read_csv('/home/jovyan/ML_Bitcoin/bitcoin_2017_to_2023.csv')

low = data['low']
high = data['high']
open = data['open']
close = data['close']
volume = data['volume']
QAV = data['quote_asset_volume']
NOT = data['number_of_trades']
TBBAV = data['taker_buy_base_asset_volume']
TBQAV = data['taker_buy_quote_asset_volume']

lowxData = [TBBAV, volume ]
midxData = [open, high, low, close, NOT]
highxData = [ QAV, TBQAV]

plt.boxplot(lowxData)
plt.title('Taker_Buy_Base_Asset_Volume x Volume')
plt.ylabel('Values')
plt.show()

plt.boxplot(midxData)
plt.title('Open x High x Low x Close x Number-Of-Trades')
plt.ylabel('Values')
plt.show()

plt.boxplot(highxData)
plt.title('Quote-Asset-Volume x Taker-Buy-Quote-Asset-Volume')
plt.ylabel('Values')
plt.show()
```
