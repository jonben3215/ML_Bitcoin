```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline 
import seaborn as sns
```

```python
dfBitcoin = pd.read_csv("bitcoin_historical_data.csv")
dfBitcoin
```

```python
# First, we have to drop all of the rows with missing values so we could calculate Q1, Q3, and IQR
# Otherwise, Q1, Q3, and IQR calculation would just show up as nan
dfBitcoin = dfBitcoin.dropna(subset=['Open'])

# Calculate the minimum value, the maximum value, the mean, the standard deviation, Q1, median, Q3, and the IQR of Open Price
minValueOpen = dfBitcoin['Open'].min()
maxValueOpen = dfBitcoin['Open'].max()
meanOpen = dfBitcoin['Open'].mean()
stdOpen = dfBitcoin['Open'].std()
q1Open = np.percentile(dfBitcoin['Open'], 25)
medianOpen = dfBitcoin['Open'].median()
q3Open = np.percentile(dfBitcoin['Open'], 75)
IQROpen = q3Open - q1Open

print("The minimum value for Open price is " + str(minValueOpen))
print("The maximum value for Open price is " + str(maxValueOpen))
print("The mean for Open price is " + str(meanOpen))
print("The standard deviation for Open price is " + str(stdOpen))
print("Q1 for Open price is " + str(q1Open))
print("The median for Open price is " + str(medianOpen))
print("Q3 for Open price is " + str(q3Open))
print("IQR for Open price is " + str(IQROpen))
```

```python
# We convert the timestamp column of the dataset to actual human readable timestamp
# As we can see, it is from 2012 to 2021
dfBitcoin['Timestamp'] = pd.to_datetime(dfBitcoin['Timestamp'], unit='s')
plt.figure(figsize=(10, 8))
plt.plot(dfBitcoin['Timestamp'], dfBitcoin['Open'], label='Open Price')
plt.title('Bitcoin Open Price Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Open Price (USD)')
plt.legend()
plt.show()
```

```python
# Below is the heat map of the correlations between each variable in the dataset
correlation_matrix = dfBitcoin[['Open', 'High', 'Low', 'Close', 'Volume_(BTC)', 'Volume_(Currency)', 'Weighted_Price']].corr()
plt.figure(figsize=(10, 8))

# Hide the top part of the map
mask = np.triu(correlation_matrix)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', mask=mask)
plt.title('The heat map of the correlations between each variable in the dataset')
plt.show()
```

As we can see throughout the closely inspection of each columns and what they do, there is a relationship between Open and High, and between Low and Close. Also, the plot graph for those 4 are very similar to one another, which indicates that there is only a small difference between each variable of Open Price, High Price, Low Price, and Close Price. Meanwhile, the Volume (BTC) and the Volume (Currency) has an opposite relationship. As we can see with the plot graph above, the Volume (BTC) is peaking on the right side of the graph and the Volume (Currency) is peaking on the left side of the graph. I also inlcuded a heat map betweeen each variable in the dataset above so that we can easily see the correlations between each variables.
